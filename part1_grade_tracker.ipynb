```python
# Task 1 - cleaning student data

raw_students = [
    {"name": "  ayesha SHARMA ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma", "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair ", "roll": "103", "marks_str": "91, 85, 88, 94, 79"}
]

data_cleaned = []

for s in raw_students:
    name = s["name"].strip().title()
    roll = int(s["roll"])
    marks = [int(x) for x in s["marks_str"].split(",")]

    print(name, roll, marks)

    data_cleaned.append((name, roll, marks))
```

```python
# Task 2 - marks analysis

subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

for i in range(len(subjects)):
    m = marks[i]

    if m >= 90:
        g = "A+"
    elif m >= 80:
        g = "A"
    elif m >= 70:
        g = "B"
    elif m >= 60:
        g = "C"
    else:
        g = "F"

    print(subjects[i], m, g)

print("Average:", sum(marks)/len(marks))
```

```python
# Task 3 - class report

class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma", [55, 68, 49, 72, 61])
]

for name, m in class_data:
    avg = sum(m)/len(m)

    if avg >= 60:
        status = "Pass"
    else:
        status = "Fail"

    print(name, avg, status)
```
